```python
# importing packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

In this Project we usedf The National UFO Reporting Center DATA  ufos.csv spreadsheet includes columns about the city, state and country where the sighting occurred, the object's shape and its latitude and longitude.

```python
# Data Loading
df = pd.read_csv('ufos.csv')
display(df.head())
```

```python
#converting data into dataFrame
df=pd.DataFrame({'Seconds':df['duration (seconds)'],'Country':df['country'],'Latitude':df['latitude'],'Longitude':df['longitude']})
df.Country.unique()
```

```python
df
```

```python
# Data cleaning
df.dropna(inplace=True)
df= df[(df['Seconds'] >= 1) & (df['Seconds'] <= 60)]
df.info()
```

```python
# categorical values to numerical values
from sklearn.preprocessing import LabelEncoder
df['country']=LabelEncoder().fit_transform(df['Country'])
```

```python
df.head()
```

```python
# Now building MODEL
# we use LOgistic REGRESSION FOR THIS DATA
from sklearn.model_selection import train_test_split
Selected_features=['Seconds','Latitude','Longitude']
X=df[Selected_features]
y=df['country']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
#now we are going to train model
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report
model=LogisticRegression()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
print(classification_report(y_test, y_pred))
print('Predicted labels: ',y_pred)
print('Accuracy: ', accuracy_score(y_test,y_pred ))
```

 Why was Logistic Regression used for this project?
 1. Classification Task
The goal is to predict a country (categorical variable) from numerical inputs.

Logistic Regression is a go-to model for multiclass classification when:

Data is linearly separable or close to it.

You want something interpretable and lightweight.

 2. Simplicity and Interpretability
Logistic Regression is fast to train, easy to understand, and provides probabilistic outputs.

In this beginner-level project, the idea is to focus on:

Cleaning data

Training a simple model

Using it in a Flask web app

 3. Baseline Model
Logistic Regression is often used as a baseline. If it performs well (in this case ~95% accuracy), you may not need complex models.

Since Latitude and Longitude are highly correlated with country, Logistic Regression can do a decent job.

Now we use Scikit-learns meta-estimators like GridSearchCV to automatically test different algorithms (models) and their hyperparameters, evaluate them, and finally pick the best-performing model?

```python
# Model selction +hyperparameter tuning
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

models = [
    ('logreg', LogisticRegression(), {'model__C': [0.1, 1, 10]}),
    ('rf', RandomForestClassifier(), {'model__n_estimators': [50, 100]}),
    ('svm', SVC(), {'model__C': [0.1, 1, 10], 'model__kernel': ['linear', 'rbf']})
]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

for name, estimator, params in models:
    pipe = Pipeline([
        ('scaler', StandardScaler()),
        ('model', estimator)
    ])

    clf = GridSearchCV(pipe, param_grid=params, cv=5, scoring='accuracy')
    clf.fit(X_train, y_train)

    print(f"Model: {name}")
    print(f"Best Score: {clf.best_score_}")
    print(f"Best Params: {clf.best_params_}")
    print("="*50)
```
